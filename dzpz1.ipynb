2. Реализовать обход графа в ширину и в глубину в виде функций на питоне. Сверить обходы с приложенными гиф файлами

```python
# Создание графа
def shortest_path(Gr, start, end):
    # Выполнение поиска минимального пути из вершины А в Б с использованием поиска в ширину
    path = nx.shortest_path(G, source=start, target=end, weight='weight')
    length = nx.shortest_path_length(G, source=start, target=end, weight='weight')
    return path, length
startt = int(input("Введите вершину А: "))
endd = int(input("Введите вершину B: "))
path, length = shortest_path(G, startt, endd)
print("Минимальный путь из вершины A в B (поиск в ширину):", path)
print("Вес пути (поиск в ширину):", length)
```

Чтобы определить, какой поиск эффективнее, можно сравнить длину найденного пути и количество шагов, необходимых для его нахождения. Обычно поиск в ширину находит кратчайший путь, но может потребовать больше памяти и времени, чем поиск в глубину. Однако, эффективность может зависеть от конкретной структуры графа.

После выполнения обоих методов поиска можно сделать вывод о том, данный способ оказался более эффективным для данного графа и конкретного пути.

3. Выполнить поиск всех путей из А в Б, вывести маршруты и их веса

```python
def find_all_paths_dfs(graph, start, end, path, paths):
    path.append(start)
    
    if start == end:
        paths.append(list(path))
    else:
        for neighbor in graph[start]:
            if neighbor not in path:
                find_all_paths_dfs(graph, neighbor, end, path, paths)
    
    path.pop()

def calculate_weight(graph, path):
    # Функция для вычисления веса маршрута
    weight = 0
    for i in range(len(path) - 1):
        # Например, если вес хранится в графе как вес ребра между вершинами path[i] и path[i+1]
        weight += graph[path[i]][path[i+1]]['weight']
    return weight


startt = int(input("Введите вершину А: "))
endd = int(input("Введите вершину Б: "))
paths = []
path = []
all_paths = find_all_paths_dfs(G, startt, endd, path, paths)

for p in paths:
    weight = calculate_weight(G, p)
    print(f'Путь: {p}, Вес: {weight}')
```
